## Data Cleaning

#### We can simply assume the next power is the same as the current power. So we use front fill method to fill NaN values.

```python
data.fillna(method='ffill', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.reset_index(inplace=True)
data.drop('index', axis=1, inplace=True)
data.head()
```

### AC with maximum power usage

```python
data.sum(axis = 0, skipna = True)
```

```python
df = pd.DataFrame(data)
print(df.columns)
```

```python
Tv ={}
for x in df.columns:
    if(x!='0'):
        Tv[x] = data[x].sum()
        
print(Tv)
print('\n')
power_max= max(zip(Tv.values(), Tv.keys()))[1]
print("AC with most usage:",power_max)
power_min= min(zip(Tv.values(), Tv.keys()))[1]
print("AC with minimum usage:",power_min)      
    
```
